```python
import os
import pickle
import tensorflow as tf
from sklearn.metrics import r2_score
```

```python
with open('/archive/bioinformatics/DLLab/KevinNguyen/data/medl_toy_data/diabetes/splits_orig.pkl', 'rb') as f:
    dictData = pickle.load(f)
```

```python
def r2(yTrue, yPred):
    # R^2 metric
    ssReg = tf.reduce_sum(tf.square(yPred - yTrue))
    ssTot = tf.reduce_sum(tf.square(yTrue - tf.reduce_mean(yTrue)))
    return 1 - (ssReg / ssTot)
```

```python
def ffnn():
    input = tf.keras.layers.Input((10,))
    dense0 = tf.keras.layers.Dense(8, activation='relu', kernel_initializer='he_normal')(input)
    dense1 = tf.keras.layers.Dense(8, activation='relu', kernel_initializer='he_normal')(dense0)
    output = tf.keras.layers.Dense(1, activation='linear', kernel_initializer='he_normal')(dense1)
    model = tf.keras.models.Model(input, output)
    model.compile(loss='mean_squared_error', optimizer='Nadam', metrics=r2)
    return model
```

```python
model = ffnn()
dfDataTrain, dfTargetTrain, _ = dictData['train0']
dfDataVal, dfTargetVal, _ = dictData['val0']
model.fit(dfDataTrain, dfTargetTrain,
          validation_data=(dfDataVal, dfTargetVal),
          epochs=100)
```
